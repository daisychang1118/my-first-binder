```python
# set: unordered, mutable, no duplicates

# create a set
myset = {1,2,3,1,2,1}
print(myset)

myset = set([1,2,3])
print(myset)

myset = set("Hello")
print(myset)

# create an empty set
myset = {} # this is wrong
print(type(myset))

myset = set()
print(type(myset))
```

```python
# 
myset = set()
myset.add(1)
myset.add(2)
myset.add(3)
myset.add(4)
myset.add(5)
myset.add(6)
print(myset)

myset.remove(3)
print(myset)

# myset.remove(9) # KeyError: 9

myset.discard(2)
print(myset)

myset.pop()
print(myset)

myset.clear()
print(myset)
```

```python
# iterate set
myset = set([1,2,3,4,5,6])

for i in myset:
    print(i)
    
# check if value in a set
if 1 in myset:
    print("yes")
```

```python
# union and intersection
odds = {1,3,5,7,9}
evens = {0,2,4,6,8}
primes = {2,3,5,7}

u = odds.union(evens)
print(u)

i = odds.intersection(evens)
print(i)

i = odds.intersection(primes)
print(i)

i = evens.intersection(primes)
print(i)

# difference between 2 sets
setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}

diff = setA.difference(setB)
print(diff)

diff = setA.symmetric_difference(setB)
print(diff)

diff = setB.difference(setA)
print(diff)

diff = setB.symmetric_difference(setA)
print(diff)
```

```python
# update set
setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}
setA.update(setB)
print(setA)


setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}
setA.intersection_update(setB)
print(setA)

setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}
setA.difference_update(setB)
print(setA)

setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}
setA.symmetric_difference_update(setB)
print(setA)
```

```python
# subset and superset
setA = {1,2,3,4,5,6}
setB = {1,2,3}
setC = {7,8}

print(setA.issubset(setB))
print(setB.issubset(setA))

print(setA.issuperset(setB))
print(setB.issuperset(setA))

print(setA.isdisjoint(setB))
print(setA.isdisjoint(setC))
```

```python
# copy set

# wrong way to copy
print("worng way to copy set")
setA = {1,2,3,4,5,6}

setB = setA
print(setB)
setB.add(7)
print(setA)
print(setB)

# way 1 to copy set
setA = {1,2,3,4,5,6}

print("way 1 to copy set")
setB = setA.copy()
print(setB)
setB.add(7)
print(setA)
print(setB)

# way 2 to copy set
setA = {1,2,3,4,5,6}

print("way 2 to copy set")
setB = set(setA)
print(setB)
setB.add(7)
print(setA)
print(setB)
```
