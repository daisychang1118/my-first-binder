```python
# dictionary: Key-Value pairs, Unordered, Mutable

# create dict
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}
mydict2 = dict(name="Mary", age=27, city="Boston")
print(mydict)
print(mydict2)

# access item
value = mydict["name"]
print(value)

# add item
mydict["city"] = "New York"
print(mydict)

# delete item
del mydict["name"]
print(mydict)

mydict.pop("age")
print(mydict)

mydict.popitem()
print(mydict)

# check if key is in dict
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}

if "name" in mydict:
    print(mydict["name"])
    
try:
    print(mydict["name"])
except:
    print("Error")
```

```python
# copy dict

# this is a wrong copy
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}
print(mydict)

mydict_cpy = mydict # this is wrong
mydict_cpy["city"] = "New York"
print(mydict)
print(mydict_cpy)

# copy way 1
print("copy dict way 1")
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}
print(mydict)

mydict_cpy = dict(mydict) # way 1
mydict_cpy["city"] = "New York"
print(mydict)
print(mydict_cpy)

# copy way 2
print("copy dict way 2")
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}
print(mydict)

mydict_cpy = mydict.copy() # way 2
mydict_cpy["city"] = "New York"
print(mydict)
print(mydict_cpy)

# merge dict
print("merge dict")
mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}
mydict2 = {"name":"Mary", "age": 22, "city": "Boston"}
mydict.update(mydict2)
print(mydict)
```

```python
# loop in dict

mydict = {"name":"Max", "age": 28, "email": "max@xyz.com"}


for key in mydict:
    print(key)
    
for key in mydict.keys():
    print(key)
    
for value in mydict.values():
    print(value)
    
for key, value in mydict.items():
    print(key, value)
```
